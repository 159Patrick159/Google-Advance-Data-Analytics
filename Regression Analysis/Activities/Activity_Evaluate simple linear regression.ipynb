### Build and fit the model

Replace the comment with the correct code. Use the variable you chose for `X` for building the model.

```python
# Seperate data for fitting
TV_data = data_sub[['Sales','TV']]
Radio_data = data_sub[['Sales','Radio']]
Social_data = data_sub[['Sales','Social_Media']]

# Define the OLS formula.

### YOUR CODE HERE ### 
TV_formula = "Sales ~ TV"
Radio_formula = "Sales ~ Radio"
Social_formula = "Sales ~ Social_Media"

# Create an OLS model.

### YOUR CODE HERE ###
TV_OLS = ols(formula=TV_formula,data=TV_data)
Radio_OLS = ols(formula=Radio_formula,data=Radio_data)
Social_OLS = ols(formula=Social_formula,data=Social_data)

# Fit the model.

### YOUR CODE HERE ### 
TV_model = TV_OLS.fit()
Radio_model = Radio_OLS.fit()
Social_model = Social_OLS.fit()

# Save the results summary.

### YOUR CODE HERE ### 
TV_summary = TV_model.summary()
Radio_summary = Radio_model.summary()
Social_summary = Social_model.summary()

# Display the model results.

### YOUR CODE HERE ### 
print("TV Model Summary:",TV_summary,'\n')
print("Radio Model Summary:",Radio_summary,'\n')
print("Social Media Model Summary:",Social_summary,'\n')
```

### Check model assumptions

To justify using simple linear regression, check that the four linear regression assumptions are not violated. These assumptions are:

* Linearity
* Independent Observations
* Normality
* Homoscedasticity

### Model assumption: Linearity

The linearity assumption requires a linear relationship between the independent and dependent variables. Check this assumption by creating a scatterplot comparing the independent variable with the dependent variable. 

Create a scatterplot comparing the X variable you selected with the dependent variable.

```python
# Create a scatterplot comparing X and Sales (Y).

### YOUR CODE HERE ###
plt.title("Linear regression model for TV sales")
sns.regplot(data=TV_data.sample(n=30,replace=True,random_state=100),y='Sales',x='TV')
plt.show()

plt.title("Linear regression model for Radio sales")
sns.regplot(data=Radio_data.sample(30,replace=True,random_state=100),y='Sales',x='Radio',color='r')
plt.show()

plt.title("Linear regression model for Social Media sales")
sns.regplot(data=Social_data.sample(30,replace=True,random_state=100),y='Sales',x='Social_Media',color='hotpink')
plt.show()
```

### Model assumption: Normality

The normality assumption states that the errors are normally distributed.

Create two plots to check this assumption:

* **Plot 1**: Histogram of the residuals
* **Plot 2**: Q-Q plot of the residuals

```python
# Calculate the residuals.
TV_res = TV_model.resid
Radio_res = Radio_model.resid
Social_res = Social_model.resid

### YOUR CODE HERE ### 

# Create a 1x2 plot figures.
fig, ax = plt.subplots(3, 2, figsize = (12,9))

# Create a histogram with the residuals. 
sns.histplot(TV_res,ax = ax[0,0])
sns.histplot(Radio_res,ax=ax[1,0],color='r')
sns.histplot(Social_res,ax=ax[2,0],color='hotpink')

### YOUR CODE HERE ### 

# Set the x label of the residual plot.
ax[0,0].set_xlabel("Residuals")
ax[1,0].set_xlabel("Residuals")
ax[2,0].set_xlabel("Residuals")

# Set the title of the residual plot.
ax[0,0].set_title("TV Residuals Distribution")
ax[1,0].set_title("Radio Residuals Distribution")
ax[2,0].set_title("Social Media Residuals Distribution")

# Create a Q-Q plot of the residuals.

### YOUR CODE HERE ### 
sm.qqplot(TV_res,ax=ax[0,1],line='s')
sm.qqplot(Radio_res,ax=ax[1,1],line='s')
sm.qqplot(Social_res,ax=ax[2,1],line='s')

# Set the title of the Q-Q plot.
ax[0,1].set_title("QQ-Plot for TV Sales")
ax[1,1].set_title("QQ-Plot for Radio Sales")
ax[2,1].set_title("QQ-Plot for Social Media Sales")

# Use matplotlib's tight_layout() function to add space between plots for a cleaner appearance.

# Show the plot.
plt.tight_layout()
```

### Model assumption: Homoscedasticity

The **homoscedasticity (constant variance) assumption** is that the residuals have a constant variance for all values of `X`.
